# Definición de funciones necesarias

```python
class News():
    def __init__( self,category, name, link, date, day, month, year, source):
        self.category = category
        self.name = name
        self.link = link
        self.date = date
        self.day = day
        self.month = month
        self.year = year
        self.source = source

    def GetKeywords(self):
        l = []
        if self.link != '':
            l += [self.link, self.link.replace(" ","")]  
        #set to avoid posible duplicates 
        return set(l)
    
    #usefull for the creation of a DataFrame later on
    def __dict__(self):
        return {
            "Categoria" : self.category,
            "Titular" : self.name,
            "Link Articulo" : self.link,
            "Fecha publicacion" : self.date,
            "Dia": self.day,
            "Mes" : self.month,
            "Año" : self.year,
            "Fuente" : self.source,
        }
    
    def __str__(self):
        return "\nCategoria: " + self.category\
                +"\nTitular: "+ self.name \
                + "\nLink Articulo: " + self.link \
                +"\nFecha publicacion: " + self.date \
                +"\nDia: " + self.day\
                +"\nMes: " + self.month\
                +"\nAño: " + self.year\
                +"\nFuente : " + self.source\
                + "\n"
```

```python
def GetDate(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content, "html.parser")
    k=soup.find_all("div", {"class": "dateTime"})
    d=""
    for p in k:
      dddate=p.find("span", {"class": "dateTime__created"}).text.replace("\n","").split("-")
      d= dddate[0]
    return d


def GetDate20(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content, "html.parser")
    k=soup.find_all("section", {"class": "article-titles"})
    d=""
    for p in k:
      dddate=p.find("span", {"class": "article-date"}).text.replace("\n","").split("-")
      d= dddate[0]
    return d
```
